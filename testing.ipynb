```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
columns = [
    'Nomor peserta', 'Nomor keluarga', 'Bobot', 'ID Kunjungan FKTP', 'Tanggal datang kunjungan FKTP',
    'Tanggal pulang kunjungan FKTP', 'Provinsi FKTP', 'Kabupaten/Kota FKTP', 'Kepemilikan FKTP', 'Jenis FKTP',
    'Tipe FKTP', 'Tingkat Pelayanan FKTP', 'Jenis Poli FKTP', 'Segmen Peserta saat akses layanan FKTP',
    'Kode dan nama diagnosis ICD 10 (3 digit)', 'Kode diagnosis ICD 10 (3 digit)', 'Kode diagnosis (beragam 3-5 digit)',
    'Nama diagnosis berasal dari kode diagnosis FKP15', 'Provinsi faskes tujuan rujukan',
    'Kabupaten/Kota faskes tujuan rujukan', 'Kepemilikan faskes tujuan rujukan', 'Jenis faskes tujuan rujukan',
    'Tipe faskes tujuan rujukan', 'Poli faskes tujuan rujukan', 'Jenis Kunjungan FKTP', 'Kelas Status Pulang Peserta'
]
df = pd.read_csv('dataset/train_fktp.txt', names = columns)
```

```python
# cek hubungan antar kolom menggunakan grafik
df.plot(x='Kode diagnosis ICD 10 (3 digit)', y='Kode diagnosis (beragam 3-5 digit)', style='o')
plt.title('Kode diagnosis ICD 10 (3 digit) vs Kode diagnosis (beragam 3-5 digit)')
plt.xlabel('Kode diagnosis ICD 10 (3 digit)')
plt.ylabel('Kode diagnosis (beragam 3-5 digit)')
plt.show()
```
